```python
%matplotlib notebook
import ipywidgets as widgets
```

```python
%matplotlib inline
newplot(plotType='TimeSeries3'); setLines(['-'])

def update_plot(VS_0, VX_0, Yield, qSmax, Ks): 
  ax1.set_title('Batch cultivation'); 
  ax1.clear(); ax2.clear() 
  ax1.set_ylim([0,20]); ax2.set_ylim([0,0.7]) 
  par(Y=Yield, qSmax=qSmax, Ks=Ks) 
  init(VS_0=VS_0, VX_0=VX_0) 
  opts['result_handling'] = 'memory' 
  opts['ncp'] = 200 # Nominal 500
  simu()

VS_0=widgets.FloatSlider(description='VS_0', value=10, min=5, max=20, step=1) 
VX_0=widgets.FloatSlider(description='VX_0', value=1.0, min=0.5, max=2.0, step=0.1) 
Yield=widgets.FloatSlider(description='Yield', value=0.5, min=0.2, max=0.7, step=0.01)
qSmax=widgets.FloatSlider(description='qSmax', value=1.0, min=0.5, max=1.5, step=0.1) 
Ks=widgets.FloatSlider(description='Ks', value=0.1, min=0.05, max=1.0, step=0.01)

widgets.interactive(update_plot, VS_0=VS_0, VX_0=VX_0, Yield=Yield, qSmax=qSmax, Ks=Ks)
```
